# 1.텍스트 크롤링

```python
import requests
import numpy as np
from bs4 import BeautifulSoup
from tqdm import tqdm
import re

base_url_1 = 'https://terms.naver.com/entry.naver?docId='
base_url_2 = '&cid=58438&categoryId=58438&expCategoryId=58438'
first_page = 3431565
last_page = 3431792

result_title = []
result_text = []

for page in tqdm(range(first_page, last_page + 1)):
    url = f"{base_url_1}{page}{base_url_2}"

    html = requests.get(url).text
    soup = BeautifulSoup(html, 'lxml')

    headword_element = soup.select_one('#content > div.section_wrap > div.headword_title > h2')

    if not headword_element:
        result_title.append(np.nan)
        result_text.append(np.nan)

    else:
        result_title.append(headword_element.get_text(strip=True))
        txt_elements = soup.select('p.txt')
        result = ""

        for element in txt_elements:
            eco_txt = element.get_text(strip=True)
            result += eco_txt + " "


        cleaned_result = re.sub(r'\t+', '', result)
        cleaned_result = re.sub(r'\n+', ' ', cleaned_result)
        cleaned_result = re.sub(r' +', ' ', cleaned_result)
        result_text.append(cleaned_result.strip())
```

```python
df_crawling = pd.DataFrame(
    columns=['대분류', '중분류', '소분류', '라벨', '제목', '텍스트'],
    data={
        '대분류': [""] * len(result_title),
        '중분류': [""] * len(result_title),
        '소분류': [""] * len(result_title),
        '라벨': [""] * len(result_title),
        '제목': result_title,
        '텍스트': result_text
    }
)
df_crawling
```

```python
df_crawling_2 = df_crawling[['제목', '텍스트']]
```

```python
from transformers import AutoTokenizer

MODEL_PATH = '/content/drive/MyDrive/절미네 가족들/1.jiwon/2.modeling/kb-albert-char-base-v2'
tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH)

def count_tokens_transformers(text):
    return len(tokenizer.encode(text))
```

```python
text = """
기업들은 생산요소를 구입하여 생산 활동을 함으로써 매출과 이익을 올리고 있다. 성공적인 경제활동을 했다면 비용보다 수입이 크게 될 것이다. 기업의 소득은 주주들에게 배당하기도 하고 새로운 사업을 위하여 직접 보유하기도 한다. 그럼에도 기업은 현재의 생산을 유지하거나 새로운 사업을 위해서 항상 자금이 필요하다. 기업은 생산물 판매가 활발하게 일어나면 시설을 확장하기 위해서 돈이 필요하고, 반대로 판매 수입이 줄어들면 부족한 돈을 빌리게 되므로 통상적으로 자금의 수요자가 된다. 가계·기업·정부 모두 수입보다 지출이 많으면 적자가 되고 반대로 수입보다 지출이 적으면 흑자가 된다. 흑자 주체는 자신들의 돈을 효과적으로 활용하기 위하여 다양한 방법으로 돈을투자하게 되고 반대로 적자 주체는 필요한 돈을 마련하기 위해서 돈을 빌리게 된다. 이와 같이 흑자 주체에서 적자 주체로 돈이 이동하는 것을 금융이라 한다. 즉 여유자금이 있는 흑자 주체(저축자: 공급자)가 돈을 빌려주거나 적자 주체(차입자: 수요자)가 돈을 빌리는 것을 금융이라 하고 금융거래가 일어나는 곳을 금융시장이라고 한다. 생산·소비·유통·분배와 같은 경제활동이 원활하게 일어나기 위해서는 각 경제주체 간에 일어나는 거래를 뒷받침할 수 있는 돈의 양이 충분해야 하고 돈의 흐름을 원활하게 해주는 금융시장이 잘 발달되어 있어야 한다. 선진국들은 생산물의 크기도 크지만 경제활동을 뒷받침하는 금융시장도 잘 발달되어 있는 데 비해서 개발도상국들은 상대적으로 생산성도 낮고 금융시장도 낙후되어 있다.
"""
print("Token count:", count_tokens_transformers(text))
```

```python
def split_into_chunks(text, max_tokens=360):
    input_ids = tokenizer.encode(text)
    chunks = []

    start_idx = 0
    while start_idx < len(input_ids):
        end_idx = start_idx + max_tokens
        # 마지막 청크의 경우 end_idx가 배열 길이를 초과할 수 있으므로 최소값을 사용
        end_idx = min(len(input_ids), end_idx)

        chunks.append(tokenizer.decode(input_ids[start_idx:end_idx]))
        start_idx = end_idx

    return chunks
```

```python
chunks = split_into_chunks(text)
for i, chunk in enumerate(chunks):
    print(f"Chunk {i + 1}:\n{chunk}\n")
```
